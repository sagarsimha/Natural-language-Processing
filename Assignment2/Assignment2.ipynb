```python
# Read text from file
with open('combinedtext.txt') as f:
        text = f.read()
```

```python
text = text.replace('@','')
text = text.replace('<p>','')
text = text.replace('\n','')
```

```python
# Tokenize(words) removing characters and punctuations
tokenized = nltk.word_tokenize(text)
```

```python
tokenized = [token.lower() for token in tokenized]
```

```python
unigrams = ngrams(tokenized, 1)
bigrams = ngrams(tokenized, 2)
trigrams = ngrams(tokenized, 3)
```

```python
unigram_freq = collections.Counter(unigrams)
bigram_freq = collections.Counter(bigrams)
trigram_freq = collections.Counter(trigrams)
```

```python
trigram_model = {}
for trigram, value in trigram_freq.items():
    trigram_model[trigram] = value / (bigram_freq.get(trigram[:-1]))
print('Done')
```

```python
sen1 = "he is from the east ."
sen2 = "she is from the east ."
sen3 = "he is from the west ."
sen4 = "she is from the west . "
sen5 = "i'll be back ."
```

```python
def prob_sentence(sentence):
    tokenized_sentence = nltk.word_tokenize(sentence)
    tokenized_sentence = [token.lower() for token in tokenized_sentence]
    
    trigrams = ngrams(tokenized_sentence, 3)
    prob_sent = 0.0
    for t in trigrams:
        if prob_sent is not None:
            prob_sent+= math.log(trigram_model.get(t))
    prob_sent = math.exp(prob_sent)
    return prob_sent
```

```python
def complete_sentence(tocomplete):
    #tocomplete = "the adventure of"
    tokenized_tocomplete = nltk.word_tokenize(tocomplete)
    tokenized_tocomplete = [token.lower() for token in tokenized_tocomplete]
    
    final_sentence = tocomplete+" "
    bi_words = tokenized_tocomplete[-2:]
    word = ""
    while(word!='.'):
        top_words = {}
        for key, value in trigram_model.items():
            if key[:-1] == tuple(bi_words):
                #print(key,value)
                top_words[key[2]] = value    
        if bool(top_words):
            word = max(top_words, key=top_words.get)
            final_sentence+= (word + " ")
            bi_words[0]=bi_words[1]
            bi_words[1]=word
        else:
            print("No words found")
            break
    print('Final sentence -', final_sentence)
```
